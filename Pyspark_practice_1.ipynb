```python
from pyspark import SparkConf,SparkContext
```

```python
if __name__=="__main__":
  conf=SparkConf().setAppName("second").setMaster("local")
  sc=SparkContext(conf=conf)
```

```python
#Two ways to create original RDD---Parallelize and by reading the text file 
#parallelize function to use create original RDD
first=sc.parallelize(['Flower Lotus',"Animal Tiger","Bird Peacock","Flag Tiranga"])
```

```python
symbols=first.map(lambda x:x.split(' ')[0])
```

```python
symbols.collect() #gives the name of elments 
```

```python
symbols.count() #count the no of rows
```

```python
values=first.map(lambda x:x.split(" ")[1])
values.collect()
```

```python
final=first.map(lambda x:(x.split(" ")[0], x.split(" ")[1]))
final.collect()
```

```python
#saving the file
final.coalesce(1).saveAsTextFile("mysecond.csv")
```

```python
type(final) #pipeline rdd because it is converted  --not original 
```

```python
dept=newRDD6.map(lambda x:x.split(",")[4]) #split used to take one particular column 
```

```python
#Filtering pf particular dept out of total
salesdept=dept.filter(lambda x:"Sales" in x)
salesdept.count()
```
